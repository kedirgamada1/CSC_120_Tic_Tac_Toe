```python
# Create empty list for each column

A = []
B = []
C = []
D = []
E = []
F = []
G = []
H = []

# Iterate through each row and append contents to correct list

for row in age_sex_dist.find_all("tr"):
    cells = row.findAll('td')
    if len(cells) == 8:
        A.append(cells[0].find(text = True).strip("\n"))
        B.append(cells[1].find(text = True).strip("\n"))
        C.append(cells[2].find(text = True).strip("\n"))
        D.append(cells[3].find(text = True).strip("\n"))
        E.append(cells[4].find(text = True).strip("\n"))
        F.append(cells[5].find(text = True).strip("\n"))
        G.append(cells[6].find(text = True).strip("\n"))
        H.append(cells[-1].find(text = True).strip("\n"))

# Print statment used to confirm data is parsing by row correctly
# print(C)
```
